##### Step 1: Install and Import Required Libraries


```python
# Install the necessary libraries
!pip install pandas
!pip install matplotlib
!pip install seaborn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here
# Drop missing values for this column
comp_data = df['ConvertedCompYearly'].dropna()

# Set the style
sns.set(style="whitegrid")

# Create the histogram and distribution plot
plt.figure(figsize=(10, 6))
sns.histplot(comp_data, kde=True, bins=40, color='skyblue')
plt.title('Distribution of Yearly Compensation (ConvertedCompYearly)')
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Number of Respondents')
plt.xlim(0, comp_data.quantile(0.95))  # Optional: limit extreme values for better visibility
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here
# Filter for full-time employed respondents
full_time = df[df['Employment'] == 'Employed, full-time']

# Drop missing compensation values
full_time_comp = full_time['ConvertedCompYearly'].dropna()

# Calculate the median compensation
median_comp = full_time_comp.median()

print(f"Median yearly compensation for full-time employed respondents: ${median_comp:,.2f}")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here
# Drop rows with missing compensation or country
df_filtered = df[['Country', 'ConvertedCompYearly']].dropna()

# Optional: Filter countries with enough respondents (e.g., 50+ entries) to avoid clutter
top_countries = df_filtered['Country'].value_counts()[df_filtered['Country'].value_counts() >= 50].index
df_filtered = df_filtered[df_filtered['Country'].isin(top_countries)]

# Set plot size and style
plt.figure(figsize=(16, 8))
sns.boxplot(data=df_filtered, x='Country', y='ConvertedCompYearly')
plt.xticks(rotation=45, ha='right')
plt.title('Yearly Compensation Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Yearly Compensation (USD)')
plt.ylim(0, df_filtered['ConvertedCompYearly'].quantile(0.95))  # Trim outliers for better visibility
plt.tight_layout()
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here
# Drop missing values in compensation column
df_clean = df[['ConvertedCompYearly']].dropna()

# Calculate IQR
Q1 = df_clean['ConvertedCompYearly'].quantile(0.25)
Q3 = df_clean['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_no_outliers = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)]

# View shape of new dataset
print(f"Original dataset shape: {df.shape}")
print(f"Refined dataset shape (without outliers): {df_no_outliers.shape}")
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here
# Select relevant columns
cols = ['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']

# Drop rows with missing values in selected columns
df_corr = df[cols].dropna()

# Convert to numeric if necessary
df_corr['WorkExp'] = pd.to_numeric(df_corr['WorkExp'], errors='coerce')
df_corr['JobSatPoints_1'] = pd.to_numeric(df_corr['JobSatPoints_1'], errors='coerce')

# Drop any newly created NaNs from conversion
df_corr = df_corr.dropna()

# Calculate correlation matrix
correlation_matrix = df_corr.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: Compensation, Experience, and Job Satisfaction')
plt.tight_layout()
plt.show()
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
## Write your code here
# Ensure correct datatypes and handle missing values
df['WorkExp'] = pd.to_numeric(df['WorkExp'], errors='coerce')
df['JobSatPoints_1'] = pd.to_numeric(df['JobSatPoints_1'], errors='coerce')

# Drop missing values from relevant columns
df_cleaned = df[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']].dropna()

# --- Scatter Plot: ConvertedCompYearly vs WorkExp ---
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_cleaned, x='WorkExp', y='ConvertedCompYearly')
plt.title('ConvertedCompYearly vs Work Experience')
plt.xlabel('Years of Work Experience')
plt.ylabel('ConvertedCompYearly')
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Scatter Plot: ConvertedCompYearly vs JobSatPoints_1 ---
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df_cleaned, x='JobSatPoints_1', y='ConvertedCompYearly')
plt.title('ConvertedCompYearly vs Job Satisfaction')
plt.xlabel('Job Satisfaction (Numeric Scale)')
plt.ylabel('ConvertedCompYearly')
plt.grid(True)
plt.tight_layout()
plt.show()
```
