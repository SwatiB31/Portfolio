### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
# Display the shape of the dataset
print("Shape of the dataset:", df.shape)

# Display column names
print("\nColumn Names:")
print(df.columns.tolist())

# Display data types and non-null counts
print("\nData Types and Non-Null Counts:")
print(df.info())

# Optional: display first few rows for a quick look
print("\nFirst 5 rows:")
print(df.head())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
# Count missing values per column
missing_values = df.isnull().sum()

# Display columns with at least one missing value
missing_values = missing_values[missing_values > 0]
print("Missing values in the dataset:\n")
print(missing_values.sort_values(ascending=False))
# Impute categorical columns with the most frequent value
df['RemoteWork'] = df['RemoteWork'].fillna(df['RemoteWork'].mode()[0])
df = df.dropna(subset=['Employment', 'JobSat'])
df = df.drop(columns=['AINextMuch less integrated', 'AINextLess integrated', 'AINextNo change', 'AINextMuch more integrated','EmbeddedAdmired'])
# Impute numerical columns with median or mean
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].median())
# Check again to confirm missing values are handled
print("\nMissing values after imputation/removal:\n")
print(df.isnull().sum().sort_values(ascending=False))
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
# Check distribution of Employment types
print("Employment Distribution:\n")
print(df['Employment'].value_counts(dropna=False))  # dropna=False includes NaN in the count

print("\nJob Satisfaction Distribution:\n")
print(df['JobSat'].value_counts(dropna=False))

print("\nYears of Professional Coding Experience Distribution:\n")
print(df['YearsCodePro'].value_counts(dropna=False).sort_index())  # sorted to show progression
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
import matplotlib.pyplot as plt

# Count the values
job_sat_counts = df['JobSat'].value_counts()

# Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(job_sat_counts, labels=job_sat_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Job Satisfaction')
plt.axis('equal')  # Equal aspect ratio ensures pie is circular
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
!pip install matplotlib-venn
from collections import Counter
import pandas as pd

# Split and count languages
def count_languages(series):
    lang_counter = Counter()
    for entry in series.dropna():
        langs = entry.split(';')
        lang_counter.update(langs)
    return lang_counter

# Count languages
have_langs = count_languages(df['LanguageHaveWorkedWith'])
want_langs = count_languages(df['LanguageWantToWorkWith'])

# Convert to DataFrame
have_df = pd.DataFrame.from_dict(have_langs, orient='index', columns=['HaveWorkedWith'])
want_df = pd.DataFrame.from_dict(want_langs, orient='index', columns=['WantToWorkWith'])

# Merge and fill missing
lang_df = have_df.join(want_df, how='outer').fillna(0)
lang_df = lang_df.astype(int).sort_values(by='HaveWorkedWith', ascending=False)

import matplotlib.pyplot as plt

# Select top 10 languages by HaveWorkedWith
top_langs = lang_df.head(10)

# Plot
top_langs.plot(kind='bar', figsize=(12, 6))
plt.title("Top Programming Languages: Have Worked With vs Want To Work With")
plt.xlabel("Programming Language")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

from matplotlib_venn import venn2

# Create sets of languages
set_have = set(have_df.index)
set_want = set(want_df.index)

# Plot Venn diagram
plt.figure(figsize=(8, 6))
venn2([set_have, set_want], set_labels=('Have Worked With', 'Want To Work With'))
plt.title("Language Usage vs Interest Overlap")
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
import pandas as pd

# Group by Country and RemoteWork, then count
remote_region_counts = df.groupby(['Country', 'RemoteWork']).size().unstack(fill_value=0)

# Optional: focus on top countries by total responses
top_countries = df['Country'].value_counts().head(10).index
remote_region_counts = remote_region_counts.loc[top_countries]

import matplotlib.pyplot as plt

# Plot grouped bar chart
remote_region_counts.plot(kind='bar', figsize=(12, 6))
plt.title('Distribution of Remote Work by Country')
plt.xlabel('Country')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.legend(title='Remote Work')
plt.tight_layout()
plt.show()

import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(remote_region_counts, annot=True, fmt='d', cmap='Blues')
plt.title('Remote Work Distribution by Country (Heatmap)')
plt.ylabel('Country')
plt.xlabel('Remote Work')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
!pip install scipy
import pandas as pd

subset = df[['YearsCodePro', 'JobSat']].copy()

# Replace values for YearsCodePro
subset['YearsCodePro'] = subset['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 51
})
subset['YearsCodePro'] = pd.to_numeric(subset['YearsCodePro'], errors='coerce')

# Map JobSat
job_sat_map = {
    'Very dissatisfied': 1,
    'Dissatisfied': 2,
    'Neither satisfied nor dissatisfied': 3,
    'Satisfied': 4,
    'Very satisfied': 5
}
subset['JobSat_Score'] = subset['JobSat'].map(job_sat_map)

# Drop rows with missing values
subset_clean = subset.dropna()

# Check the number of rows
print(f"Valid rows: {len(subset_clean)}")

from scipy.stats import pearsonr, spearmanr

# Pearson correlation
pearson_corr, pearson_p = pearsonr(subset['YearsCodePro'], subset['JobSat_Score'])

# Spearman correlation
spearman_corr, spearman_p = spearmanr(subset['YearsCodePro'], subset['JobSat_Score'])

print(f"Pearson Correlation: {pearson_corr:.3f} (p={pearson_p:.3f})")
print(f"Spearman Correlation: {spearman_corr:.3f} (p={spearman_p:.3f})")
```
